1. Convert the timestamp column to datetime object.
2. Create columns with hour, day, week.
3. What day of the week had the highest activity?

First, I convert the timestamp column to datetime using, pd.to_datetime()

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

Control that timestamp is converted:

```python
df
```

```python
df.timestamp.dtype
```

The timestamp is converted, now let's extract hour, day, and week.

```python
df['hour'] = df['timestamp'].dt.hour
df['day'] = df['timestamp'].dt.day_name()
df['week'] = df['timestamp'].dt.isocalendar().week
```

```python
df
```

Now, let's show what day of the week had the highest activity. This would be the day that has more clicks.

```python
df.groupby('day')['clicks'].sum().sort_values()
```

Tuesday had the highest activity, with 13 clicks.
